```python
# ðŸ›’ FMCG Sales Forecasting - Data Preparation

import pandas as pd
import os

# Path to dataset
data_path = "/kaggle/input/fmcg-daily-sales-data-to-2022-2024/"

# Load the main dataset
df = pd.read_csv(os.path.join(data_path, "FMCG_2022_2024.csv"))

print("Raw shape:", df.shape)
print(df.head())
print(df.columns)

# --- Convert date column ---
df['date'] = pd.to_datetime(df['date'])

# --- Create revenue column ---
df['revenue'] = df['units_sold'] * df['price_unit']

# --- Aggregate to daily revenue ---
daily = (
    df.groupby('date')['revenue']
    .sum()
    .reset_index()
    .rename(columns={'revenue': 'daily_revenue'})
)

print("âœ… Daily revenue time series created. Shape:", daily.shape)
print(daily.head())

# --- Save processed file ---
os.makedirs("/kaggle/working/data", exist_ok=True)
daily.to_csv("/kaggle/working/data/daily_sales.csv", index=False)

print("ðŸ’¾ Saved processed file: /kaggle/working/data/daily_sales.csv")
```
